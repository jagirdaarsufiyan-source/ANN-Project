```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle 
```

```python
data = pd.read_csv('student_lifestyle_100k.csv')
data.head()
```

```python
print(data.columns)
```

```python
label_encoder_Study_Hours = LabelEncoder()
data['Study_Hours'] = label_encoder_Study_Hours.fit_transform(data['Study_Hours'])
data.head() 
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo=OneHotEncoder()
geo_encoder=onehot_encoder_geo.fit_transform(data[['Stress_Level']])
geo_encoder
```

```python
onehot_encoder_geo.get_feature_names_out(['Stress_Level'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder.toarray(), columns=onehot_encoder_geo.get_feature_names_out(['Stress_Level']))
geo_encoded_df.head()
```

```python
data = pd.concat([data.drop('Stress_Level', axis=1), geo_encoded_df], axis=1)
data.head()
```

```python
with open('label_encoder_Study_Hours.pkl', 'wb') as file:
    pickle.dump(label_encoder_Study_Hours, file)

with open('onehot_encoder_geo.pkl', 'wb') as file:
    pickle.dump(onehot_encoder_geo, file)
```
